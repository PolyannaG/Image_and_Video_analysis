```python
#We want the first frame of the new video
#video = cv2.VideoCapture('output.mp4')
#ret, frame = video.read()
#frame = data.checkerboard()

cap = cv2.VideoCapture("output.mp4")

ret, image = cap.read()
```

```python
def CornerDetector(img,maxCorners,qualityLevel,minDistance,blockSize,useHarrisDetector, plot=True):
    params = dict( maxCorners = maxCorners,
                       qualityLevel = qualityLevel,
                       minDistance = minDistance,
                       blockSize = blockSize,
                       useHarrisDetector = useHarrisDetector)
    
    image_ret=np.copy(img)
    gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    corners = cv2.goodFeaturesToTrack(gray, mask = None, **params)
    for i in corners:
        x,y = i.ravel()
        cv2.circle(image_ret,(x,y),3,255,-1)
    if plot:
        if useHarrisDetector:
            plt.title('Corner detection using Harris Detector')
        else:
            plt.title('Corner detection using Shi-Tomasi Detector')
        plt.imshow(image_ret, cmap='gray')  
    return image_ret  
```

```python
fig, axs = plt.subplots(3, 2,figsize=(30, 30))

qualityLevel = 0.7
frame1=CornerDetector(image, maxCorners = 100,qualityLevel=qualityLevel,minDistance = 7,blockSize = 7,useHarrisDetector = False,plot=False)
axs[0,0].imshow(frame1)
axs[0,0].set_title('qualityLevel={}'.format(qualityLevel), fontsize=14)

qualityLevel = 0.5
frame2=CornerDetector(image, maxCorners = 100,qualityLevel=qualityLevel,minDistance = 7,blockSize = 7,useHarrisDetector = False,plot=False)
axs[0,1].imshow(frame2)
axs[0,1].set_title('qualityLevel={}'.format(qualityLevel), fontsize=14)

qualityLevel = 0.25
frame3=CornerDetector(image, maxCorners = 100,qualityLevel=qualityLevel,minDistance = 7,blockSize = 7,useHarrisDetector = False,plot=False)
axs[1,0].imshow(frame3)
axs[1,0].set_title('qualityLevel={}'.format(qualityLevel), fontsize=14)

qualityLevel = 0.1
frame4=CornerDetector(image, maxCorners = 100,qualityLevel=qualityLevel,minDistance = 7,blockSize = 7,useHarrisDetector = False,plot=False)
axs[1,1].imshow(frame4)
axs[1,1].set_title('qualityLevel={}'.format(qualityLevel), fontsize=14)

qualityLevel = 0.02
frame3=CornerDetector(image, maxCorners = 100,qualityLevel=qualityLevel,minDistance = 7,blockSize = 7,useHarrisDetector = False,plot=False)
axs[2,0].imshow(frame3)
axs[2,0].set_title('qualityLevel={}'.format(qualityLevel), fontsize=14)

qualityLevel = 0.005
frame4=CornerDetector(image, maxCorners = 100,qualityLevel=qualityLevel,minDistance = 7,blockSize = 7,useHarrisDetector = False,plot=False)
axs[2,1].imshow(frame4)
axs[2,1].set_title('qualityLevel={}'.format(qualityLevel), fontsize=14)

fig.suptitle('Corner Detection using Shi-Tomasi Detector, changing qualityLevel', fontsize=40)
```
